```python
#import libraries
import pandas as pd
import openpyxl
import os

#setup pwc
os.chdir(r'C:\Users\sshocc\Documents\PythonChatGPT\training') 
os. getcwd()
#read excel file
aled = pd.read_excel(r"Assets, Liabilities, External, Deposits 1 EX.xlsx")
key = pd.read_excel(r"Key Business Reasons EX.xlsx")
```

```python
print(aled)
print(key)
```

```python
# Names of the columns from 'key' DataFrame to merge based on the conditions
reason_column = 'Column3'
source_name_column = 'Source.Name'  # replace with the actual name if different

# Perform the merge operation, now including the 'Source.Name' column
result_df = aled.merge(key[[reason_column, source_name_column, 'Current', 'MoM Change']], 
                       left_on=['Value', 'Preceding Change'], 
                       right_on=['Current', 'MoM Change'], 
                       how='left')

# Copy the reason and source name from 'key' to 'aled' where the conditions match
result_df['Appended Reason'] = result_df[reason_column]
result_df['Appended Source Name'] = result_df[source_name_column]

# Drop the helper columns from the 'key' DataFrame
result_df = result_df.drop(columns=[reason_column, source_name_column, 'Current', 'MoM Change'])

# Save the result back to a new Excel file
result_df.to_excel('aled_updated.xlsx', index=False)  # replace with your desired save path
```
